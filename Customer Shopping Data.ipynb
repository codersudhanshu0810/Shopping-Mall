# Importing Necessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Data

```python
df = pd.read_csv('customer_shopping_data.csv')
df
```

```python
df.columns = np.char.title(list(df.columns))
df
```

# isnull()

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Invoice_No'] = df['Invoice_No'].str.replace('I', '')
df['Customer_Id'] = df['Customer_Id'].str.replace('C', '')
df
```

```python
df.info()
```

# Type Conversion

```python
df['Invoice_No'] = df['Invoice_No'].astype(int)
df['Customer_Id'] = df['Customer_Id'].astype(int)
df['Invoice_Date'] = df['Invoice_Date'].str.replace('/', '-')
df
```

```python
df['Invoice_Date'] = pd.to_datetime(df['Invoice_Date'], format='%d-%m-%Y')
df.info()
```

```python
df['Amount'] = df['Quantity'] * df['Price']
df
```

# Value Counts

```python
gdf = df['Gender'].value_counts()
gdf = gdf.reset_index()
gdf
```

# Data Visualization

```python
plt.figure(figsize = (10, 5))
plt.bar(gdf['Gender'], gdf['count'])
plt.title('Shopping according to Gender')
plt.xlabel('Gender')
plt.ylabel('Gender Count')
plt.savefig('Shopping according to Gender.png', format='png')
```

# Groupby

```python
ndf = df.groupby('Gender').agg(Earning_by_Gender = ('Amount', 'sum'))
ndf = ndf.reset_index().sort_values(by='Earning_by_Gender', ascending=False)
ndf
```

```python
ndf.info()
```

```python
plt.figure(figsize = (10, 5))
plt.bar(ndf['Gender'], ndf['Earning_by_Gender'])
plt.title('Earning according to Gender')
plt.xlabel('Gender')
plt.ylabel('Earning')
plt.savefig('Earning according to Gender.png', format='png')
```

# Groupby

```python
kdf = df.groupby('Category').agg(Earning_by_Category = ('Amount', 'sum'))
kdf = kdf.reset_index().sort_values(by='Earning_by_Category', ascending=False)
kdf
```

# Data Visualization

```python
plt.figure(figsize = (18, 5))
plt.bar(kdf['Category'], kdf['Earning_by_Category'])
plt.title('Earning according to Category')
plt.xlabel('Category')
plt.ylabel('Earning')
plt.savefig('Earning according to Category.png', format='png')
```

# Groupby

```python
idf = df.groupby('Payment_Method').agg(PaymentMethod = ('Amount', 'sum'))
idf = idf.reset_index().sort_values(by = 'PaymentMethod', ascending=False)
idf
```

# Data Visualization

# Groupby

```python
jdf = df.groupby('Shopping_Mall').agg(Shopping_Mart = ('Amount', 'sum'))
jdf = jdf.reset_index().sort_values(by = 'Shopping_Mart', ascending=False)
jdf
```

```python
df['Month Name'] = df['Invoice_Date'].dt.month_name()
df
```

```python
rdf = df.groupby('Month Name').agg(HighestSales = ('Amount', 'sum'))
rdf = rdf.reset_index().sort_values(by='HighestSales', ascending=False)
rdf
```

# Data Visualization

```python
plt.figure(figsize = (18, 5))
plt.bar(rdf['Month Name'], rdf['HighestSales'])
plt.title('Earning in Month')
plt.xlabel('Month Name')
plt.ylabel('Amount')
plt.savefig('Earning in Shopping Month.png', format='png')
```
